```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    # PDF 문서를 열고 각 페이지의 텍스트를 추출합니다.
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

if __name__ == "__main__":
    # 변환할 PDF 파일 경로 설정
    pdf_path ="/content/drive/MyDrive/25.03.21 AI 시스템 구축 관련 참고자료-1/업무편람 예시/정보시스템 재해복구지침.PDF"
    # 텍스트 추출
    extracted_text = extract_text_from_pdf(pdf_path)

    # 추출된 텍스트를 UTF-8 인코딩으로 파일에 저장
    with open("/content/drive/MyDrive/25.03.21 AI 시스템 구축 관련 참고자료-1/업무편람 예시/정보시스템_output.txt", "w", encoding="utf-8") as f:
        f.write(extracted_text)

    print("PDF 변환 완료: output.txt 파일에 저장되었습니다.")


```
